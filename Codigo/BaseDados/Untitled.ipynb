```python
import pandas as pd
from siuba import *
```

```python
df = pd.read_excel("../../BaseDados/Microdados/Pré_Cactus.xlsx", sheet_name='Sheet1')
```

```python
df.columns
```

```python
df =(df >>
    gather("ano", "IDEB_9o", "n2013", "n2015"))
```

```python
df["ano"] = df["ano"].replace('n2013', 2013)
df["ano"] = df["ano"].replace('n2015', 2015)
```

```python
df1 = pd.read_excel("../../BaseDados/Microdados/Pré_Cactus.xlsx", sheet_name='Sheet2')
df1 =(df1 >>
    gather("ano", "IDEB_5o", "a2013", "a2015"))
df1["ano"] = df1["ano"].replace('a2013', 2013)
df1["ano"] = df1["ano"].replace('a2015', 2015)
```

```python
df_merge_col = pd.merge(df, df1, on=['ano', 'Município'])
df_merge_col.to_excel("../../BaseDados/Microdados/Pré_Cactus.xlsx")
```
